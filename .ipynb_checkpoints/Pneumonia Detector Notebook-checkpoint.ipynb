```python
# set up:
%matplotlib inline
import random
import matplotlib.pylab as plt

from PIL import Image
import numpy as np
from scipy.misc import imresize
import glob

# from vgg16 import VGG16
import keras.backend as K
import tensorflow as tf
from keras.models import Model
from keras.callbacks import History 
from keras.optimizers import Adam
from keras.layers import Dense, Dropout, Flatten, Input, BatchNormalization
from keras.preprocessing.image import ImageDataGenerator
from keras.utils.np_utils import to_categorical
```

```python
# preprocess images

# squash all images to size (256,256)
# adapted from https://github.com/ostegm/resizing/blob/master/Resizing.ipynb
size = 256
dpath = './chest_xray/'
list_X = []
list_y = []
for data_group in ['test', 'val', 'train']:
    for classified in ['NORMAL', 'PNEUMONIA']: 
        fnames = list(glob.iglob(dpath + f'{data_group}/{classified}/*.jpeg'))
        for fname in fnames:
            y = 0 if classified == 'NORMAL' else 1
            list_y.append(y)
            img = Image.open(fname).convert("L")
            img_arr = np.array(img)
            resized = imresize(img_arr, (size, size))
            list_X.append(resized)

# shuffle data
combined = list(zip(list_X, list_y))
random.shuffle(combined)
list_X[:], list_y[:] = zip(*combined)
X = np.stack(list_X, axis=0)
y = np.stack(list_y, axis=0)

# divide into training, validation, and test data
X_train = X[:len(X)//2,:,:]
y_train = y[:len(X)//2]
X_val = X[(len(X)//2):int(3/4*(len(X))),:,:]
y_val = y[len(X)//2:int(3/4*(len(X)))]
X_test = X[int(3/4*(len(X))):,:,:]
y_test = y[int(3/4*(len(X))):]
print('Train data shape: ', X_train.shape)
print('Train labels shape: ', y_train.shape)
print('Validation data shape: ', X_val.shape)
print('Validation labels shape: ', y_val.shape)
print('Test data shape: ', X_test.shape)
print('Test labels shape: ', y_test.shape)
```
